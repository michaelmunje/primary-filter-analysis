```python
import pandas as pd
df_arti = pd.read_csv('phase1-multicoll-artifact.csv')
df_arti.head()
```

```python
index = df_arti.iloc[[168, 1281, 454]].index
```

```python
df_arti.drop(index, inplace=True)
```

```python
df_miscal = pd.read_csv('phase1-multicoll-miscal.csv')
df_miscal.head()
```

```python
index = df_miscal.iloc[[13, 38, 76, 79, 779]].index
```

```python
df_miscal.drop(index, inplace=True)
```

```python
df_miscal['Output'] = df_miscal['Output'] * 2
```

```python
df_feat = pd.read_csv('phase1-multicoll-featureless.csv')
df_feat.head()
```

```python
index = df_feat.iloc[[19, 34, 43, 55, 722, 16, 724, 49]].index
```

```python
df_feat.drop(index, inplace=True)
```

```python
df_feat['Output'] = df_feat['Output'] * 3
```

```python
df_dust = pd.read_csv('phase1-multicoll-dust.csv')
df_dust.head()
```

```python
index = df_dust.iloc[[350, 392, 1094, 1313, 350, 1291]].index
```

```python
df_dust.drop(index, inplace=True)
df_dust['Output'] = df_dust['Output'] * 4
```

```python
df = pd.concat([df_arti, df_miscal, df_feat, df_dust])
df = df.drop_duplicates(subset='Filename')
df['Output'] = df['Output'].astype(int)
df.to_csv('phase2-res.csv', index=False)
```
