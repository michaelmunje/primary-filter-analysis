```python
import pandas as pd
df = pd.read_csv('phase2-res.csv')

feats_df = df.drop(columns=['Filename', 'Output'])
Y = df['Output']
df.head()
```

```python
import statsmodels.api as sm
from statsmodels.genmod.generalized_linear_model import GLM
from statsmodels.genmod import families
import statsmodels.stats.tests.test_influence

X2 = sm.add_constant(feats_df.values)

model = sm.MNLogit(Y, X2)
fit = model.fit(method='newton', maxiter=500)
print(fit.summary())
```

```python
import numpy as np

def forward_selection(df):
    all_features = df.columns
    current_features = []
    
    x2 = np.ones(len(df))
    model = sm.MNLogit(Y, x2)
    fit = model.fit(method='newton', maxiter=500)
    current_aic = fit.aic
    
    for i in range(len(all_features)):
        print("Step " + str(i+1))
        aics = []
        for j in range(len(all_features)):
            if all_features[j] in current_features:
                aics.append(1000000000000000000)
                continue
            x = feats_df[current_features + [all_features[j]]].values
            model = sm.MNLogit(Y, sm.add_constant(x))
            fit = model.fit(method='newton', maxiter=500)
            aics.append(fit.aic)
        if min(aics) >= current_aic:
            break
        else:
            current_features.append(all_features[np.argmin(aics)])
            current_aic = min(aics)
    return current_features
```

```python
features = forward_selection(feats_df)
```

```python
out_df = feats_df[features].copy()
out_df['Output'] = df['Output']
out_df['Filename'] = df['Filename']
out_df.to_csv('phase3-var-sel.csv', index=False)
```
